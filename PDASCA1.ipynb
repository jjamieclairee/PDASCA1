## Step 5: Clean the Data

In this step, we perform various data cleaning tasks to ensure that the dataset is ready for analysis. These tasks include removing duplicates, handling missing values, ensuring correct data types for columns, and standardizing certain columns.

#### (a) Remove duplicates

```python
merged_df = merged_df.drop_duplicates()
```

Explanation: The `.drop_duplicates()` function removes any rows that are identical across all columns, ensuring we only have unique records in the dataset.

#### (b) Drop rows with null values - write

Next, we drop rows that contain null values in critical columns (i.e., `max_salary`, `pay_period`, and `location`). These columns are essential for our analysis, and any rows missing data in these columns should be excluded.

```python
critical_columns = ['max_salary', 'pay_period', 'location']
merged_df.dropna(subset=critical_columns, inplace=True)
```

Explanation: We define `critical_columns` as the columns that are crucial for our analysis. The `.dropna()` function removes any rows that contain missing (`NaN`) values in these specified columns.

#### (c) Ensure salary columns are numeric

To ensure the integrity of the salary data, we check that the `max_salary` and `min_salary` columns are numeric. If any non-numeric values exist, they are coerced to `NaN`.

```python
merged_df['max_salary'] = pd.to_numeric(merged_df['max_salary'], errors='coerce')
merged_df['min_salary'] = pd.to_numeric(merged_df['min_salary'], errors='coerce')
```

Explanation: The `.to_numeric()` function is used to convert the `max_salary` and `min_salary columns` to numeric values. The `errors='coerce'` argument ensures that any non-numeric values are replaced with `NaN`, which helps maintain data consistency.

#### (d) Clean and standardise the location and pay_period columns

To further clean the data, we standardize the `location` and `pay_period` columns. This step ensures that values in these columns follow consistent formatting.

```python
merged_df['location'] = merged_df['location'].str.strip().str.title()
merged_df['pay_period'] = merged_df['pay_period'].str.lower().str.strip()
```

Explanation:
- Location: We use `.str.strip()` to remove any leading or trailing whitespace and `.str.title()` to ensure that the location names are properly capitalized (e.g., "new york" becomes "New York").
- Pay Period: Similarly, `.str.strip()` removes any excess spaces, and `.str.title()` ensures the pay periods are standardized (e.g., "hourly" becomes "Hourly").

#### (e) Convert Original Listed Time to Datetime and Group Data by Month

Lastly, we ensure the `original_listed_time column` is in a proper datetime format. This allows us to group the data by month and calculate the average salary trends over time.

```python
# Convert 'original_listed_time' to datetime if it isn't already
merged_df['original_listed_time'] = pd.to_datetime(merged_df['original_listed_time'], errors='coerce')

# Group data by month and calculate the average salary (using 'max_salary' or 'med_salary')
salary_trends = merged_df.groupby(merged_df['original_listed_time'].dt.to_period('M'))['max_salary'].mean().reset_index()

# Convert 'Period' to 'timestamp' for plotting
salary_trends['original_listed_time'] = salary_trends['original_listed_time'].dt.to_timestamp()

# Verify the grouped data
print(salary_trends.head())
```

Explanation:
- Datetime Conversion: We use `pd.to_datetime()` to ensure that the `original_listed_time` column is in the correct datetime format. This is necessary for time-based analysis.
- Group by Month: The `.dt.to_period('M')` function is used to group the data by month, and we calculate the average `max_salary` for each month using `.mean()`.
- Convert Period to Timestamp: After grouping by month, we convert the period back to a timestamp for better compatibility with plotting functions.

**Conclusion**
This data cleaning process ensures that we have a consistent and clean dataset, ready for further analysis. By removing duplicates, handling missing values, ensuring numeric consistency, and standardizing key columns, we ensure that our analysis results are accurate and meaningful.

## Step 6: Standardise Salaries

In this step, we perform the standardization of salaries by ensuring that all salaries are presented in a consistent format (yearly salary). Additionally, we visualize the distribution of job pay periods (yearly, hourly, etc.) and clean up any invalid salary values.

#### (a) Visualisation of Proportion of Pay Periods using Bar Chart

- Bar Chart Creation: The plot(kind='bar') function creates the bar chart. We customize the colors for each bar using a list of color codes.
- Labels and Title: We add labels for both the x-axis (Pay Period) and y-axis (Number of Jobs), and provide a title to the plot.

#### (b) Standardisation

To standardize the salary data, we need to convert different pay periods into yearly salary equivalents. For instance, hourly salaries are converted into yearly salaries based on the average number of hours worked in a year, and monthly salaries are adjusted to reflect the yearly equivalent.

```python
# Define a function to standardize salaries
def standardize_salary(row):
    if row['pay_period'] == 'hourly':
        return row['max_salary'] * 2080  # Hourly to yearly
    elif row['pay_period'] == 'monthly':
        return row['max_salary'] * 12  # Monthly to yearly
    elif row['pay_period'] == 'weekly':
        return row['max_salary'] * 52  # Weekly to yearly
    elif row['pay_period'] == 'yearly':
        return row['max_salary']  # Already yearly
    else:
        return None  # Handle undefined pay periods
```

```python
# Apply standardization and drop invalid rows
merged_df['yearly_salary'] = merged_df.apply(standardize_salary, axis=1)
merged_df.dropna(subset=['yearly_salary'], inplace=True)
```

**Explanation:**
- Standardization Function: The function `standardize_salary(row)` is defined to convert salaries from different pay periods to yearly equivalents:
    - Hourly: Multiplies by 2080 (assuming 40 hours per week for 52 weeks).
    - Monthly: Multiplies by 12 to get the yearly equivalent.
    - Weekly: Multiplies by 52 to get the yearly equivalent.
    - Already Yearly: No conversion needed.
- Apply Function: The `apply()` function is used to apply the `standardize_salary()` function to each row in the dataset, creating a new column for yearly salary.
- Drop Invalid Rows: We drop rows that contain invalid yearly salary values (e.g., if the salary could not be standardized).

#### (c) Ensure no negative or zero salaries

After standardizing the salaries, we ensure that there are no negative or zero values for salaries. These invalid salary values are removed from the dataset to maintain data integrity.

```python
merged_df = merged_df[merged_df['yearly_salary'] > 0]
```

**Explanation:**
- Filtering Invalid Salaries: We use the condition `merged_df['yearly_salary'] > 0` to filter out any rows where the yearly salary is less than or equal to zero, ensuring that only valid salaries remain in the dataset.

**Conclusion:**

By performing the tasks in Step 6, we ensure that the pay periods are standardized to yearly equivalents, the data is clean, and invalid salary values are removed. The bar chart provides a visual representation of how job postings are distributed across different pay periods, and the salary standardization ensures that we have consistent salary data for further analysis.

## Step 7: Handling Outliers

#### Use IQR to filter extreme outliers

```python
Q1 = merged_df['yearly_salary'].quantile(0.25)
Q3 = merged_df['yearly_salary'].quantile(0.75)
IQR = Q3 - Q1
merged_df = merged_df[(merged_df['yearly_salary'] >= Q1 - 1.5 * IQR) & 
                      (merged_df['yearly_salary'] <= Q3 + 1.5 * IQR)]
```

### Why Remove Outliers in This Dataset?

Outliers are data points that significantly differ from other observations in the dataset. These extreme values can skew statistical analysis, model performance, and lead to misleading conclusions. In this dataset, which contains job listings and salary information, it is particularly important to remove outliers for the following reasons:

#### 1. Ensuring Salary Representations are Realistic:
- **Job salaries** can vary greatly depending on the role, company, location, and work type. However, there may be extreme salaries (outliers) that do not represent typical job postings, such as:
  - Extremely high-paying executive roles.
  - Outliers caused by data entry errors (e.g., salaries that are incorrectly typed as 1,000,000+).
- **Removing outliers** helps ensure that the salary data accurately reflects the typical job market and job posting trends, preventing unrealistic salaries from distorting the analysis.

#### 2. Improving Model Performance:
- **Machine learning models** and statistical analysis (e.g., regression) are often sensitive to outliers, which can disproportionately affect model accuracy. For example:
  - If you're using salary data to predict job role suitability or salary forecasting, outliers may lead to **skewed predictions** or result in a poor model fit.
- **By removing outliers**, your model will focus on typical job postings and salary trends, which are more representative of the general population.

#### 3. Better Salary Analysis:
- **Analyzing salary distributions** (e.g., comparing salaries for different work types or locations) is more meaningful without the distortion of extreme outliers.
- **Visualizations** (e.g., histograms, bar plots) of salaries can be more accurately interpreted if the dataset is not skewed by a few extreme salaries.
- **Comparing average salaries** across different regions or job roles will be more reliable when you are not including outlier salaries that could significantly shift the mean.

#### 4. Detecting Data Entry Errors:
- Sometimes, **outliers are caused by data entry errors** (e.g., missing data replaced by a very high number or incorrect data). Removing outliers helps ensure that **incorrect or anomalous data points** don’t impact your analysis.
- **By cleaning the dataset**, you reduce the chance of basing conclusions on faulty or erroneous data.

#### 5. Focus on "Normal" Salary Distributions:
- For most job roles, **salaries tend to follow certain patterns** within specific ranges (e.g., software engineers in a particular region or industry will generally fall within a similar salary range).
- Removing extreme outliers ensures that your analysis focuses on the "core" distribution of salaries rather than being skewed by extreme examples that don’t represent typical hiring trends.

---

### **How to Remove Outliers:**

The **Interquartile Range (IQR)** method is used to filter out extreme values that fall outside the acceptable range. This ensures that only salaries within a reasonable range are included in your analysis.

- **Outliers are defined as any value that lies below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.**
- **By removing these values**, you ensure that only data within the "typical" range is analyzed, allowing for more accurate conclusions.

---

### **Conclusion:**
Removing outliers from your dataset improves the **reliability** of your analysis by:
1. Ensuring **realistic salary distributions**.
2. Improving the accuracy and relevance of **models and predictions**.
3. Allowing for better visualizations of salary trends.
4. Reducing the impact of **data entry errors** or anomalies.

By doing so, you focus on the most common, representative data points, leading to more insightful and actionable conclusions about the job market.

#### 4. Scatter Plot: Salary vs. Job Popularity
